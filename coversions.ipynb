```python
import re
```

```python
mp3_file=r'/content/ar.audio.mp3'
from pydub import AudioSegment
from pydub.playback import play
audio=AudioSegment.from_mp3(mp3_file)
audio.export('ar.wav',format="wav")
```

```python
r = sr.Recognizer()
```

```python
with sr.AudioFile('/content/ar.wav') as source:  # Replace with your file name
    audio = r.record(source)
```

```python

text = r.recognize_google(audio)
print("Recognized Text:", text)
```

```python
text = text.lower()
text = text.replace('plus', '+')
text = text.replace('minus', '-')
text = text.replace('times', '*')
text = text.replace('multiplied by', '*')
text = text.replace('divided by', '/')
```

```python
# This regex extracts expressions like 8+4/2 or 5*3-1
expr = re.findall(r'[\d\+\-\/\.]+(?:\s[\+\-\/]\s[\d\+\-\/\.]+)', text)

```

```python
if expr:
    expression = expr[0].strip()
    print("Extracted Expression:", expression)

    try:
        result = eval(expression)
        print("Result:", result)
    except Exception as e:
        print("Evaluation Error:", e)
else:
    print("No valid expression found.")
```
